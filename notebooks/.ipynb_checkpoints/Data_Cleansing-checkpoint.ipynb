```python
import os
import cv2
import numpy as np
import pandas as pd
from pprint import pprint
```

```python
data_ver = "augmented_2"
data_path = os.path.join("..", "data", data_ver)
image_path = os.path.join(data_path, "train_imgs")
data_df = pd.read_csv(os.path.join(data_path, data_ver + ".csv"))
data_df = data_df.drop_duplicates(["image"])
```

```python
image_list = os.listdir(image_path)
print(len(image_list))
```

```python
data_df
```

```python
for name in data_df["image"]:
    if name not in image_list:
        print(f"{name} doesn't exist in image dir.")
```

```python
keypoint_labels = data_df.columns[1:].to_list()

remove_file_list = []
remove_row_idx_list = []

for idx, row in enumerate(data_df.to_numpy()):
    file_name = row[0]
    keypoints = row[1:]
    if any(keypoints < 0):
        remove_file_list.append(file_name)
        remove_row_idx_list.append(idx)
```

```python
pprint(remove_file_list)
```

```python
print(len(remove_file_list))
print(len(remove_row_idx_list))
```

```python
for idx, (file_name, row_idx) in enumerate(zip(remove_file_list, remove_row_idx_list)):
    if file_name != data_df.iloc[row_idx, 0]:
        print(f"Not matched at {row_idx}")
        print(file_name)
        print(data_df.iloc[row_idx, 0])
```

```python
for file_name in remove_file_list:
    file_path = os.path.join(image_path, file_name)
    if os.path.isfile(file_path):
        os.remove(file_path)
    else:
        print(f"{file_name} doesn't exist.")
print(f"After removing, the number of image files: {len(os.listdir(image_path))}")

data_df.drop(index=remove_row_idx_list, axis=0, inplace=True)
print(f"After removing, the length of dataframe: {len(data_df)}")
```

```python
data_df.to_csv(os.path.join(os.path.join(data_path, data_ver + ".csv")), index=False)
```
